```r
cgctable <- read.csv('~/jackgl/Census_allThu Sep 13 17_30_17 2018.csv')
genes_brain <- read.csv('~/jackgl/analyses/tcga_gene_lists/frequently_mutated_genes_TCGA_brain.csv')
genes_lung <- read.csv('~/jackgl/analyses/tcga_gene_lists/frequently_mutated_genes_TCGA_lung.csv')
genes_colon <- read.csv('~/jackgl/analyses/tcga_gene_lists/frequently_mutated_genes_TCGA_colon.csv')
```

```r
cancer_type <- 'colon'
```

```r
if (cancer_type=='lung') {
    toMatch <- c('lung')
    tcga_table <- genes_lung 
} else if (cancer_type=='colon') {
    toMatch <- c('colon','colorectal')
    tcga_table <- genes_colon
} else if (cancer_type=='brain') {
    toMatch <- c('astrocytoma','meningioma','head','oligodendroglio')
    tcga_table <- genes_brain
}


# Get genes affecting chosen cancer type
cgctable_filt<-cgctable[grep(paste(toMatch,collapse='|'),cgctable$Tumour.Types.Somatic),]
# Filter for only Tier 1 genes
cgctable_filt<-cgctable_filt[cgctable_filt$Tier==1,]
# Filter for TSGs
cgctable_filt<-cgctable_filt[grep('TSG',cgctable_filt$Role.in.Cancer),]

cat(paste('# Genes only in CGC:',length(which(!(cgctable_filt$Gene.Symbol %in% tcga_table$Symbol))),'\n'))
cat(paste('# Genes only in TCGA:',length(which(!(tcga_table$Symbol %in% cgctable_filt$Gene.Symbol))),'\n'))
cat(paste('# Genes in both:',length(which(tcga_table$Symbol %in% cgctable_filt$Gene.Symbol)),'\n'))

cgctable_filt

tcga_table
```
