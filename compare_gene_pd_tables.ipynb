```r
load('./data/gene_pd_allsites.rda')
gene_pd_allsites <- gene_pd_m
load('./data/gene_pd_exon.rda')
gene_pd_exons <- gene_pd_m
load('./data/genes.rda')

gene_pd_allsites <- gene_pd_allsites[order(gene_pd_allsites$gene_id),]
gene_pd_exons <- gene_pd_exons[order(gene_pd_exons$gene_id),]
genes <- genes[order(genes$gene_id),]
#--- Check that the rows of each table correspond to same genes
identical(genes$gene_id,gene_pd_allsites$gene_id)
identical(genes$gene_id,gene_pd_exons$gene_id)

gene_pd_allsites$gene_sym <- genes$gene_sym
gene_pd_exons$gene_sym <- genes$gene_sym
```

```r
par(mar=c(10,5,5,5))
options(repr.plot.width=9, repr.plot.height=5)
barplot(gene_pd_allsites$p[order(-gene_pd_allsites$p)][1:30],names=gene_pd_allsites$gene_sym[order(-gene_pd_allsites$p)][1:30],las=2)
```

```r
par(mar=c(10,5,5,5))
barplot(gene_pd_exons$p[order(-gene_pd_exons$p)][1:30],names=gene_pd_exons$gene_sym[order(-gene_pd_exons$p)][1:30],las=2)
options(repr.plot.width=9, repr.plot.height=5)
```
